```python
#  matplotlib을 이용하여 구간 [0,2]에서 함수 f(x) = x^2 + sin(3x+7) 함수의 그래프를 그리세요. 

import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,2, 0.02)
y = x**2 + np.sin(3*x+7)

plt.plot(x,y)
plt.show()
```
